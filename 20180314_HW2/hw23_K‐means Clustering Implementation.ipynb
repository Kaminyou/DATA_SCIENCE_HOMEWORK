```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data_ori = pd.read_csv("cdata.csv")
df = data_ori.copy()
#定義一個函數來算距離
def se(x,y,cx,cy):
    return (x-cx)**2 + (y-cy)**2
```

```python
k = 4
se_temp = 0
stop = 1
round_nb = 0
list_temp = []
#從所有點找出四個隨機的點做成DataFrame
centroid = df.loc[np.random.choice(df.index,k),["x","y"]]
#把這個DataFrame中的index去掉
centroid.reset_index(drop = True, inplace = True)
while stop >= 1:
    #統計多少次
    round_nb = round_nb + 1 
    #做前一個隨機點的備分
    centroid_pre = centroid.copy()
    #判斷句離哪個點近後將之與該點存成list，把分組出來的結果存回原先的資料分群row
    for index in df.index:
        for k_run in range(k):
            se_temp = se(df.iloc[index,0],df.iloc[index,1],centroid.iloc[k_run,0],centroid.iloc[k_run,1])
            list_temp.append([se_temp, k_run])
        list_temp.sort()
        df.iloc[index,2] = (list_temp[0][1])
        list_temp.clear()
    #做一個以分群名的group
    now_group = df.groupby("cluster")
    #為之後要變成英文做準備
    df["cluster"] = df["cluster"].astype(np.int8)
    #更新四個判斷點，直接透過cluster的分群算出其所對應所有xy的平均
    for index in centroid.index:
        centroid.loc[index] = df[df["cluster"] == index][["x", "y"]].mean()
    #如果新的4點組成dataframe與原先相等則停止迴圈
    if centroid.equals(centroid_pre):
        stop = 0
    #否則進行並畫圖
    else:
        for name, group in now_group:
            #轉成英文
            name=chr(name + 65)
            plt.scatter(group.x, group.y, label = name)
        plt.scatter(centroid["x"], centroid["y"], marker="x", color = "k", label="Central")
        plt.xlabel("x", fontsize=20)
        plt.ylabel("y", fontsize=20, rotation=0)
        plt.legend(loc = "best")
        plt.title("Round" + str(round_nb), fontsize=15)
        plt.show()                                                  
```

```python
#由k=2開始
k = 2
list_k = []
list_se = []
#至k=10，裡面演算之方式與第一題一樣
while k <= 10:
    df2 = data_ori.copy()
    centroid = df2.loc[np.random.choice(df2.index,k),["x","y"]]
    centroid.reset_index(drop = True, inplace = True)
    se_temp = 0
    stop = 1
    list_temp = []
    while stop >= 1:   
        centroid_pre = centroid.copy()
        for index in df2.index:
            for k_run in range(k):
                se_temp = se(df2.iloc[index,0],df2.iloc[index,1],centroid.iloc[k_run,0],centroid.iloc[k_run,1])
                list_temp.append([se_temp, k_run])
            list_temp.sort()
            df2.iloc[index,2] = (list_temp[0][1])
            list_temp.clear()
        now_group = df2.groupby("cluster")
        df2["cluster"] = df2["cluster"].astype(np.int8)
        for index in centroid.index:
            centroid.loc[index] = df2[df2["cluster"] == index][["x", "y"]].mean()
        if centroid.equals(centroid_pre):
            stop = 0
    total_se = 0
    total_se_temp = 0
    #得到這次確定的四個分群與中心點後開始計算sse並存成list
    for index in df2.index:
        total_se_temp = se(df2.iloc[index,0],df2.iloc[index,1],centroid.iloc[df2.iloc[index,2],0],centroid.iloc[df2.iloc[index,2],1])
        total_se = total_se + total_se_temp
    list_k.append(k)
    list_se.append(total_se)
    k = k + 1
    del(df2)
    list_temp.clear()
#最後用這些list來畫圖
plt.plot(list_k, list_se)
plt.xlabel("Number of k", fontsize=10)
plt.ylabel("Sum of square error", fontsize=10)
plt.show()
```
