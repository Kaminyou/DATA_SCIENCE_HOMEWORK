```python
#User‐based Collaborative Filtering (U‐CF)
#Cosine Similarity Metric
def U_CF_cos(matrix_original,epsilon = 1e-9): #epsilon is to avoid many problem of the value of 0
    matrix_original_Ucos = np.copy(matrix_original)
    for i in range(matrix_original_Ucos.shape[0]):
        mean_i = np.mean(matrix_original_Ucos[i,].nonzero())
        for j in range(matrix_original_Ucos.shape[1]):
            if matrix_original_Ucos[i,j] != 0:
                matrix_original_Ucos[i,j] = matrix_original_Ucos[i,j] - mean_i
                if matrix_original_Ucos[i,j] == 0:
                    matrix_original_Ucos[i,j] = matrix_original_Ucos[i,j] + (1e-9) #avoid the task of mask different meaning of 0
                else:
                    pass
            else:
                pass
    sim = matrix_original_Ucos.dot(matrix_original_Ucos.T) + epsilon #avoid divide 0
    norms = np.array([np.sqrt(np.diagonal(sim))])
    matrix_rating = (sim / norms / norms.T)
    return matrix_rating

#User‐based Collaborative Filtering (U‐CF)
#Pearson Correlation Coefficient  
def U_CF_pcc(matrix_original):
    matrix_original_Upcc = np.copy(matrix_original)
    with np.errstate(divide='ignore', invalid='ignore'):
        matrix_rating = np.corrcoef(matrix_original_Upcc)
    matrix_rating = np.nan_to_num(matrix_rating)
    return matrix_rating


#Item‐based Collaborative Filtering (I‐CF)
#Cosine Similarity Metric
def I_CF_cos(matrix_original,epsilon = 1e-9):
    matrix_original_Icos = np.copy(matrix_original)
    matrix_original_Icos = matrix_original_Icos.T
    for i in range(matrix_original_Icos.shape[0]):
        mean_i = np.mean(matrix_original_Icos[i,].nonzero())
        for j in range(matrix_original_Icos.shape[1]):
            if matrix_original_Icos[i,j] != 0:
                matrix_original_Icos[i,j] = matrix_original_Icos[i,j] - mean_i
                if matrix_original_Icos[i,j] == 0:
                    matrix_original_Icos[i,j] = matrix_original_Icos[i,j] + (1e-9) #avoid the task of mask different meaning of 0
                else:
                    pass
            else:
                pass
    sim = matrix_original_Icos.dot(matrix_original_Icos.T) + epsilon #avoid divide 0
    norms = np.array([np.sqrt(np.diagonal(sim))]) 
    matrix_rating = (sim / norms / norms.T)
    return matrix_rating


#Item‐based Collaborative Filtering (I‐CF)
#Pearson Correlation Coefficient    
def I_CF_pcc(matrix_original):
    matrix_original_Ipcc = np.copy(matrix_original)
    matrix_original_Ipcc = matrix_original_Ipcc.T
    with np.errstate(divide='ignore', invalid='ignore'):
        matrix_rating = np.corrcoef(matrix_original_Ipcc)
    matrix_rating = np.nan_to_num(matrix_rating)
    return matrix_rating

def predict_for_movie(matrix_original,matrix_rating, method="user", k=10):
    #matrix_rating_sort = np.argsort(matrix_rating,axis=-1)
    matrix_original_pre = np.copy(matrix_original)
    if method == "user":
        matrix_prediction = np.copy(matrix_original_pre)
        for i in range(matrix_original_pre.shape[0]):
            k_user = np.argsort(matrix_rating[i,:])[-k:][::-1]      #largest k index
            for j in range(matrix_original_pre.shape[1]): 
                if matrix_original_pre[i,j] == 0:
                    prediction = (matrix_rating[i,k_user].dot(matrix_original_pre[k_user,j]))/np.sum(matrix_rating[i,k_user])
                    matrix_prediction[i,j] = prediction
                else:
                    pass
        matrix_prediction = np.nan_to_num(matrix_prediction)
    if method == "item":
        matrix_prediction = np.copy(matrix_original_pre.T)
        matrix_original_trans = np.copy(matrix_original_pre.T)
        for i in range(matrix_original_trans.shape[0]):
            k_item = np.argsort(matrix_rating[i,:])[-k:][::-1]      #largest k index
            for j in range(matrix_original_trans.shape[1]): 
                if matrix_original_trans[i,j] == 0:
                    prediction = (matrix_rating[i,k_item].dot(matrix_original_trans[k_item,j]))/np.sum(matrix_rating[i,k_item])
                    matrix_prediction[i,j] = prediction
                else:
                    pass
        matrix_prediction = matrix_prediction.T
        matrix_prediction = np.nan_to_num(matrix_prediction)
    matrix_prediction[matrix_prediction > 5] = 5
    matrix_prediction[matrix_prediction < 1] = 1
    return matrix_prediction                

def RMSE(matrix_prediction, matrix_real):
    from sklearn.metrics import mean_squared_error
    matrix_prediction_nonzero = matrix_prediction[np.nonzero(matrix_real)].flatten()  #only nonzero values which mean that were not ranked before are condisered
    matrix_real_nonzero = matrix_real[np.nonzero(matrix_real)].flatten()
    mse = mean_squared_error(matrix_prediction_nonzero,matrix_real_nonzero)
    rmse = np.power(mse, (1/2))
    return rmse
```

## Part 2: Matrix Factorization without and with Bias

### Load user rating file, covert the rating file into user-item matrix

```python
import numpy as np
import pandas as pd

ratingfile = 'ratings.data'
header = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv(ratingfile, sep='\t', names=header)

n_users = df.user_id.unique().shape[0]
n_items = df.item_id.unique().shape[0]
print('Number of users = ' + str(n_users) + ' | Number of movies = ' + str(n_items))

user_item_matrix = np.zeros((n_users, n_items))
for line in df.itertuples():
    user_item_matrix[line[1] - 1, line[2] - 1] = line[3]
```

### Split the user-item matrix set into training set and testing set

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# split the data set into training set and testing set
train_data, test_data = train_test_split(df, test_size=0.25)

# transfer the train set and test set of ratings to user-item matrix
    
train = pd.DataFrame(train_data)
test = pd.DataFrame(test_data)

train_matrix = np.zeros((n_users, n_items))
for line in train.itertuples():
    train_matrix[int(line[1]) - 1, int(line[2]) - 1] = line[3]

test_matrix = np.zeros((n_users, n_items))
for line in test.itertuples():
    test_matrix[int(line[1]) - 1, int(line[2]) - 1] = line[3]  
```

### Compute the Rooted mean square error to evaluate the accuaracy

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(pred_rate, true_rate):
    prediction = pred_rate[true_rate.nonzero()].flatten() 
    true_rate = true_rate[true_rate.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, true_rate))
```

### Get the index for the non-zero rating in training and testing user-item matrix

```python
# Index matrix for training data
train_ind = train_matrix.copy()
train_ind[train_ind > 0] = 1
train_ind[train_ind == 0] = 0

# Index matrix for test data
test_ind = test_matrix.copy()
test_ind[test_ind > 0] = 1
test_ind[test_ind == 0] = 0
```

### Prediction function

```python
# Predict the unknown ratings through the dot product of the latent features for users and items and the bias
def prediction_sgd_bias(P, Q, b_global, b_user, b_item):
    ratings = np.dot(P, Q.T)
    for u in range(len(b_user)):
        for i in range(len(b_item)):
            ratings[u][i] = b_global +  b_user[u] + b_item[i] + ratings[u][i]
    ratings = np.nan_to_num(ratings)
    return ratings
```

### Rooted mean square error computation function

```python
# Calculate the RMSE for SGD (without bias)
def sgd_rmse(ind, data_matrix, P, Q):
    return np.sqrt(np.sum((ind * (data_matrix - np.dot(P, Q.T)))**2)/len(data_matrix[data_matrix > 0]))

# Calculate the RMSE for SGD with bias
def sgd_bias_rmse(ind, data_matrix, P, Q, b_global, b_user, b_item):
    return np.sqrt(np.sum((ind * (data_matrix - prediction_sgd_bias(P, Q, b_global, b_user, b_item)))**2)/len(data_matrix[data_matrix > 0]))
```

### Stochastic Gradient Descent (SGD) function

```python
def sgd(train, test, train_ind, test_ind, factor_k, alpha, lmbda, iteration):
    print ("Computing using SGD ......")
    
    train_errors = []
    test_errors = []
    
    m, n = train.shape # Number of users and items in training set
    
    # initialize Latent user feature matrix and Latent movie feature matrix
    P = 3 * np.random.rand(m, factor_k)  # Latent user feature matrix
    Q = 3 * np.random.rand(n, factor_k)  # Latent movie feature matrix
    
    # Only consider non-zero elements in the training matrix
    users, items = train.nonzero()
    for it in range(iteration):
        print("SGD Iteration", it)
        for u, i in zip(users, items):
            e = train[u, i] - (np.dot(P[u], Q[i].T))  # Calculate error for gradient
            P[u] += alpha * (e * Q[i] - lmbda * P[u])  # Update latent user feature matrix
            Q[i] += alpha * (e * P[u] - lmbda * Q[i])  # Update latent movie feature matrix
        train_rmse = sgd_rmse(train_ind, train, P, Q)  # Calculate root mean squared error from train dataset
        test_rmse = sgd_rmse(test_ind, test, P, Q)  # Calculate root mean squared error from test dataset
        train_errors.append(train_rmse)
        test_errors.append(test_rmse)

    return train_errors, test_errors
```

### Matrix Factoriztion with Bias (SGD)

```python
def sgd_bias(train, test, train_ind, test_ind, factor_k, alpha, lmbda, iteration):
    print ("Computing using SGD ......")
    
    train_errors = []
    test_errors = []
    
    m, n = train.shape # Number of users and items in training set
    train_negative = train.copy()
    
    # initialize Latent user feature matrix and Latent movie feature matrix
    P = 3 * np.random.rand(m, factor_k)  # Latent user feature matrix
    Q = 3 * np.random.rand(n, factor_k)  # Latent movie feature matrix
    
    # initialize global bias
    b_global = np.mean(train[np.nonzero(train)]) 
    
    # initialize user bias
    b_user = []
    
    for user_m in np.arange(m):
        b_user_m = np.mean(train[user_m,np.nonzero(train[user_m,:])])
        b_user.append(b_user_m)
    b_user = np.array(b_user)
    b_user = np.nan_to_num(b_user)
    
    # initialize item bias
    b_item = []
    for item_n in np.arange(n):
        b_item_n = np.mean(train[np.nonzero(train[:,item_n]),item_n])
        b_item.append(b_item_n)
    b_item = np.array(b_item)
    b_item = np.nan_to_num(b_item)
    
    # Only consider non-zero elements in the training matrix
    users, items = train.nonzero()
    for it in range(iteration):
        print("SGD Iteration", it)
        for u, i in zip(users, items):
            e = train[u, i] - (np.dot(P[u], Q[i].T)) - b_user[u] - b_item[i] - b_global # Calculate error for gradient
            P[u] += alpha * (e * Q[i] - lmbda * P[u])  # Update latent user feature matrix
            Q[i] += alpha * (e * P[u] - lmbda * Q[i])  # Update latent movie feature matrix
            b_user[u] += alpha * (e - lmbda * b_user[u]) # Update user bias matrix
            b_item[i] += alpha * (e - lmbda * b_item[i]) # Update item bias matrix
            b_global += alpha * (e - lmbda * b_global) # Update global bias matrix

        train_rmse = sgd_bias_rmse(train_ind, train, P, Q, b_global, b_user, b_item) # Calculate root mean squared error from train dataset
        test_rmse = sgd_bias_rmse(test_ind, test, P, Q, b_global, b_user, b_item)  # Calculate root mean squared error from test dataset
        train_errors.append(train_rmse)
        test_errors.append(test_rmse)

    return train_errors, test_errors
```
